```python
# generate HD vectors

# params
N = 10000
p = 0.1
k = 0.25 * p
Q = 1000

hd = [None] * Q

# ref vector
hd[0] = np.random.choice([0, 1], size=N, p=[1 - p, p])

for q in range(1, Q):
    y = np.random.choice([0, 1], size=N, p=[k, 1 - k])
    z = np.random.choice([0, 1], size=N, p=[1 - (p * k / (1 - p + k * p )), p * k / (1 - p + k * p )])
    hd[q] = np.bitwise_or(np.bitwise_and(hd[q - 1], y), z)
#     print(np.count_nonzero(hd[0] == hd[q]))
```

```python
# generate HD role vectors for features

hd_role = [None] * data.shape[1]

for i in range(data.shape[1]):
    hd_role[i] = np.random.choice([0, 1], size=N, p=[1 - p, p])
```

```python
# helper functions

def hd_bind(x, y):
    # CDT for sparse binary distributed representations, Rachkovskij
    
    z = np.bitwise_or(x, y)
    zk = np.zeros(x.shape[0], dtype=int)
    
    # use suitable K to maintain sparsity - depending on p
    # QUESTION how to choose a good K ?
#     K = 25
#     for k in range(K):
#         zk = np.bitwise_or(zk, np.random.permutation(z))
        
    # keep "sparse-ness"
    while True:
        zk = np.bitwise_or(zk, np.bitwise_and(z, np.random.permutation(z)))
        if np.count_nonzero(zk) >= (p * N):
            return zk
        
#     return np.bitwise_and(z, zk)

def get_hd(a):
    # blank start
    v = np.zeros(N, dtype=int)
    
    for i in range(2700, data.shape[1]):
#         r = data[:, i].max() - data[:, i].min()
        q = np.digitize(a[i], np.histogram(data[:, i], bins=Q)[1][:-1]) - 1
    
#         try:
        hd_f = hd[q]
        hd_r = hd_role[i]
#         except IndexError:
#             print(i)
#             print(a[i])
#             print(np.histogram(data[:, i], bins=Q))
#             print(q)
#             histo = np.histogram(data[:, i], bins=Q)
            
        
        v = np.bitwise_or(v, hd_bind(hd_f, hd_r))
    
    return hd_bind(v, v)
```
